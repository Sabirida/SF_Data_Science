### PostgreSQL: Установка и подключение

НЕМНОГО О POSTGRES
В курсе мы для удобства работы с данными использовали Metabase. Естественно, это не единственный доступный инструмент, в частности, мы неоднократно упоминали систему управления базами данных PostgreSQL. Пришла пора ближе познакомиться с этим инструментом.

PostgreSQL — это свободная система управления реляционными базами данных (СУРБД, или RDBMS) с открытым исходным кодом.

Она поддерживает такие сущности как транзакции, схемы и внешние ключи и часто позиционируется как СУБД, наиболее строго соответствующая стандартам SQL. К её преимуществам перед другими — коммерческими и некоммерческими — относят также более высокий уровень безопасности.

В данном модуле PostgreSQL понадобится нам в качестве инструмента, связующего базу данных и механизмы языка Python.

ЗАПУСК СЕРВЕРА
Теперь вы можете запустить сервер базы данных — database server.

 Зачем это нужно? 
Сервер — это программа, которая работает в фоновом режиме и ожидает входящие команды.

Когда команды поступают, сервер обрабатывает их и выдаёт ответ. Если сервер не запустить, доступ к базе организовать не получится, так как запросы будет некуда отправлять.

Есть два варианта запустить сервер: 

С помощью командной строки.
С помощью графического интерфейса pgAdmin.
Все дальнейшие действия мы будем рассматривать для каждого из вариантов.

Запустим сервер с помощью командной строки.

Для этого нужно в терминале (командной строке) вызвать следующую команду. Рабочую директорию менять не надо, так как указывается абсолютный путь к папке.

 Запуск на macOS
pg_ctl -D /usr/local/var/postgres start 
 Запуск на Linux
sudo service postgresql start
 Запуск на Windows
Перейдите в директорию C:\Program Files\PostgreSQL\11\bin, затем выполните в командной строке:

pg_ctl.exe -D "C:\Program Files\PostgreSQL\11\data" start
В данном случае ключ -D указывает на директорию, в которую по умолчанию устанавливается сервер postgres.

Если вы работаете на Windows (или если при установке вы указали путь, отличающийся от стандартного), возможно, вам придётся поменять путь, который вы указываете после ключа -D.

Например, у вас может быть другая версия postgres и вместо 11 вы будете писать 11.7.

Параметр start сообщает, что надо запустить сервер.

Если вы хотите остановить сервер, то параметр start надо заменить на stop, если перезапустить — на restart.

Если сервер не запускается...
 Порядок действий
Для начала проверьте, правильно ли указана версия postgres в вызовах командной строки.
Затем посмотрите на текст ошибки. Если вы видите что-то вроде: Возможно, порт 5432 занят другим процессом — это значит, что порт, используемый по умолчанию, уже занят и нужно запускать сервер на другом порте.
 Запускаем сервер на другом порте
Нужно перед параметром start в команде указать ключ -p и другой порт (просто выберите любое другое число):

pg_ctl.exe -D "C:\Program Files\PostgreSQL\11\data" -p 5431 start
В случае неудачи выберите порт снова и запустите повторно. Повторяйте действия, пока не получится.

Если всё прошло успешно, нужно установить пользователя. 

Для этого необходимо в той же папке, где запускали сервер, прописать psql -U postgres.

postgres — это пользователь по умолчанию, который создаётся, когда вы устанавливаете PostgreSQL, пароль данного пользователя — это пароль, введённый при установке.

После ввода пароля командная строка станет такой postgres=#.

Затем можно создать пользователя с именем вашего пользователя (имя должно быть как в системе) через команду

CREATE USER имя пользователя WITH PASSWORD 'пароль'; 
выведется фраза CREATE ROLE.

Также можно задать пользователю статус суперпользователя.

Суперпользователь может переопределить все ограничения доступа в базе данных.

Этот статус несёт опасность и назначать его следует только в случае необходимости.

Создать нового суперпользователя может только суперпользователь.

Чтобы задать статус, необходимо прописать команду

ALTER ROLE имя пользователя WITH SUPERUSER;
или

ALTER ROLE имя пользователя WITH CREATEDB CREATEROLE;

Итак, синтаксис создания таблиц следующий:

CREATE TABLE table_name (

 Что есть что?
→ table_name — название таблицы, по которому мы будем к ней обращаться;

→ type — тип данных в столбце;

→ column_constraint — ограничение на столбец;

→ table_constraint — ограничение на всю таблицу.

Названия таблицы и столбца — очевидные параметры.

Например, названием таблицы может быть Students, а названием одного из столбцов — Name.

Дальше мы рассмотрим остальные параметры, а именно типы данных и ограничения — на столбец и на таблицу.

ОГРАНИЧЕНИЯ
Теперь кратко осветим тему ограничений.

 Ограничение на столбец
column_constraint в коде выше — ограничение на столбец.

Какие бывают ограничения:

NOT NULL — значения в столбце не могут быть пропусками (NULL).
UNIQUE — значения в столбце должны быть уникальными.
Помните, что пропусков (NULL) всё равно может быть несколько, так как каждый из них считается уникальным значением.

PRIMARY KEY — первичный ключ; это ограничение является сочетанием NOT NULL и UNIQUE. С помощью этого метода вы можете создать внешний ключ из одного столбца. Иначе придётся использовать ограничение на таблицу (см. ниже).
CHECK — проверка условия при вставке данных в таблицу или их обновлении.
Например, если у вас есть столбец price, вы можете задать условие, которое будет проверять, что цена имеет положительное значение.

REFERENCES — внешний ключ; это ограничение позволяет обозначить, что значение столбца ссылается на другой столбец.
Важно! На столбец можно можно накладывать сразу несколько ограничений.

 Ограничение на таблицу
table_constraint в коде выше обозначает ограничение на всю таблицу.

Какие бывают ограничения:

UNIQUE(column_list) — то же, что UNIQUE для столбца, но можно передать несколько столбцов через запятую.
PRIMARY KEY(column_list) — то же, что PRIMARY KEY для столбца, но можно создать составной ключ, передав несколько столбцов через запятую.
CHECK (condition) — то же, что CHECK для столбца, но условие может затрагивать несколько столбцов в таблице.
REFERENCES — то же, что REFERENCES для столбца.
Теперь мы располагаем всеми необходимыми знаниями для создания таблицы в PostgreSQL.

import psycopg2

try:
    connection = psycopg2.connect(
        user="postgres",
        password="ваш_пароль",

        database="sf_test",
    )
    cursor = connection.cursor()

        connection.close()
        print("PostgreSQL connection is closed")

Транзакции
 Добавить страницу в мои закладки
ОБЩИЕ СВЕДЕНИЯ
Транзакция — это неделимая (атомарная) операция, которую вы можете произвести с базой.

При этом транзакция может содержать в себе несколько операций.

Например, вы можете вставить несколько строк в базу и всё это будет считаться единой транзакцией. В случае если произошла ошибка и только часть строк была вставлена в базу, транзакция «откатится», то есть все внесённые изменения будут отменены.

Когда мы выполняем несколько операций в коде, в конце нужно сделать коммит (от англ. commit), то есть показать системе, что мы должны выполнить предыдущие операции как единое целое. 

Каноничный пример транзакции в реальном мире — перевод денег с одного счёта на другой в банке. 

Как происходит перевод:

Система снимает деньги с одного счёта.
Система добавляет деньги на другой счёт.
Если бы операция перевода не являлась транзакцией, то в случае ошибки между первым и вторым этапом деньги пропали бы безвозвратно.

На самом деле, дело не только в защите от ошибок при удалении, но и, например, в защите от неконсистентности данных.

Транзакции подчиняются принципу ACID:

Atomic (атомарность) — если произошёл коммит, значит, все операции были выполнены.
Consistent (консистентность) — можно сделать коммит, только если он удовлетворяет всем ограничениям (например, NOT NULL).
Isolated (изолированность) — снаружи транзакции не видно, что происходит внутри, пока не произошёл коммит.
Durable (долговечность) — независимо от ошибок в системе, транзакция, которая была совершена (то есть был коммит) останется совершённой, даже если сразу после коммита сервер упадёт.
ПРИМЕР ТРАНЗАКЦИИ
Сделаем новую таблицу в виде транзакции:

```python
try:
    connection = psycopg2.connect(
        user="postgres",
        password="260599Da.postgres",
        host="127.0.0.1",
        port="5432",
        database="sf_test",
    )
    cursor = connection.cursor()

    create_table_query = '''CREATE TABLE mobile
          (ID INT PRIMARY KEY     NOT NULL, 
          MODEL           TEXT    NOT NULL, 
          PRICE         REAL); '''

    cursor.execute(create_table_query)
    connection.commit()
    print("Table created successfully in PostgreSQL")
except (Exception, psycopg2.DatabaseError) as error:
    print("Error while creating PostgreSQL table", error)
finally:
    if connection:
        cursor.close()
        connection.close()
        print("PostgreSQL connection is closed")
```

Продвинутые операции CRUD

В psycopg2 можно выполнить любой запрос сразу для нескольких строк.

Для этого есть специальный оператор:

cursor.executemany(query, records)

где query — это запрос, который будет выполнен несколько раз; records — список, состоящий из данных, которые мы хотим использовать в запросе.

ВСТАВКА НЕСКОЛЬКИХ СТРОК
Применим этот метод для вставки нескольких строк.

```python
try:
    connection = psycopg2.connect(
        user="postgres",
        password="260599Da.postgres",
        host="127.0.0.1",
        port="5432",
        database="sf_test",
    )
    cursor = connection.cursor()

    # Удалим все данные (без условия)  
    cursor.execute("DELETE FROM mobile")
  
    # Теперь вставим строки в таблицу
    sql_insert_query = """ INSERT INTO mobile (id, model, price) VALUES (%s,%s,%s) """  
    records_to_insert = [  
        (1, 'Iphone XS', 1050),  
     (2, 'Nutella', 30),  
     (3, 'Pixel', 970)  
    ]  
    result = cursor.executemany(sql_insert_query, records_to_insert)  
    connection.commit()  

except (Exception, psycopg2.Error) as error:
    if connection:
        print("Error", error)
finally:
    if connection:
        cursor.close()
        connection.close()
```

Работа с данными на диске

ЧТЕНИЕ ИЗ ФАЙЛА
Один из способов добавить данные в базу — прочитать их из файла.

Psycopg2 позволяет сделать это с помощью метода cursor.copy_from().

Давайте рассмотрим весь порядок действий на примере.

Допустим, у нас есть файл mobile.csv со следующими данными:

7,nokia,100  
8,blackberry,500 

Для дальнейшей работы скопируйте эти данные в любой текстовый редактор и сохраните полученный файл под именем mobile.csv.

Давайте прочитаем этот файл и запишем данные из него в нашу базу:

Мы получили то, что ожидали увидеть: данные после прошлых частей курса остались на месте и к ним добавились новые данные из файла.

```python
try:
    connection = psycopg2.connect(
        user="postgres",
        password="260599Da.postgres",
        host="127.0.0.1",
        port="5432",
        database="sf_test",
    )
    cursor = connection.cursor()

    with open("mobile.csv", 'r') as f:  
        cursor.copy_from(f, 'mobile', sep=',')  
    connection.commit()  

except (Exception, psycopg2.Error) as error:
    if connection:
        print("Error", error)
finally:
    if connection:
        cursor.close()
        connection.close()
```
