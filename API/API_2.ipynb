### Яндекс.Метрики

Яндекс.Метрика — это система веб-аналитики вашего сайта.  Она собирает данные о посещаемости вашего сайта и обобщает их в удобном для восприятия виде. 

Принцип работы следующий. Посетители взаимодействуют со страницами сайта, на которых установлен счётчик. Код счётчика исполняется и передаёт сервису Яндекс.Метрика данные как о самой странице, так и событиях, произошедших при взаимодействии посетителя с ней.

Сбор данных о посещаемости сайта и о том, что делают его пользователи, позволяет измерять ключевые показатели, чтобы затем улучшать их, тем самым повышая прибыльность бизнеса или, в случае некоммерческих сайтов, доставлять информацию, полезную большему числу пользователей.

Таким образом, данные веб-аналитики помогают принимать правильные решения о развитии сайта. 

Основная единица измерения в Яндекс.Метрике — это визит.

Визит — последовательность действий (активность) одного посетителя на сайте (на одном счётчике). 

Активностью посетителя считаются:
* просмотры страниц,
* переходы по внешним ссылкам,
* загрузки файлов. 
Визит завершён, если активность отсутствует в течение некоторого времени. По умолчанию это время равняется 30 минутам, но можно задать и другое время в настройках счётчика.

Яндекс.Метрика (и её API) оперирует двумя основными сущностями: группировки (англ. dimensions) и метрики (англ. metrics).

Эта терминология пришла из технологии обработки данных OLAP. Она также встречается и в другой системе веб-аналитики — уже знакомой вам Google Аналитике.

Группировки (англ. Dimensions) — это какой-либо атрибут визита или действия.

Как ясно из названия, группировки позволяют объединять данные по определённым признакам.

Например, мы можем сгруппировать визиты по источнику, из которого пришли посетители, по типу устройства, с которого пользователи просматривали сайт, по дате посещения и т. п.

Яндекс.Метрика позволяет использовать в одном запросе несколько группировок. В этом случае в параметрах запроса они указываются через запятую. В следующих блоках этого модуля мы  научимся создавать такие запросы.

Метрики (англ. Metrics) — это численные величины, которые отображают сводные показатели визитов или действий.

Метрики характеризуют совокупный показатель по определённой группе данных.

Например, если у нас установлена группировка Браузер посетителя и метрика Визиты, то в отчёте мы увидим количество визитов, совершённых из разных браузеров.

Методика подсчёта метрики зависит от её типа.

Например, при выборе метрики Визиты мы получим сумму визитов по каждой группе показателей, указанных в группировке. Выбрав метрику Время на сайте мы увидим для каждого значения группировки среднее время, которое пользователи провели на сайте, рассчитанное как сумма длительностей всех визитов, делённая на число визитов.

Среди метрик также — количество уникальных посетителей, показатель отказов, средний доход визита, доля мужчин или женщин среди посетителей сайта.

### Параметры запросов

Запрос к API Яндекс.Метрики представляет собой GET-запрос к адресу https://api-metrika.yandex.net/stat/v1/data с набором параметров. 

В этом блоке рассмотрим набор параметров запроса, которые определяют критерии получаемых из Яндекс.Метрики данных.

Итак, мы будем использовать следующие параметры запроса:
* metrics — метрики, то есть количественные показатели, характеризующие посетителей сайта;
* dimensions — группировки, определяющие, по каким критериям все посетители сайта будут объединены в группы;
* date1 — начало периода, данные по которому будут помещены в отчёт;
* date2 — окончание периода, данные по которому будут помещены в отчёт; 
* limit — максимально допустимое количество строк в отчёте;
* offset — номер строки отчёта, с которой должна начаться выгрузка данных;
* ids — номер счётчика, установленный на сайте.

#### 1. Параметр metrics

В параметре metrics нужно перечислить список метрик (максимальное количество в запросе — 20).

Важно! Это обязательный параметр, то есть без указания значения для этого параметра запрос выполняться не будет.

    ym:s:visits — количество визитов;
    ym:s:pageviews — суммарное количество просмотров страниц;
    ym:s:users — количество уникальных посетителей (за отчётный период);
    ym:s:bounceRate — показатель отказов;
    ym:s:pageDepth — глубина просмотра;
    ym:s:avgVisitDurationSeconds — средняя продолжительность визитов в секундах
В одном запросе можно использовать несколько метрик. В этом случае они перечисляются в одну строку через запятую без пробелов. Вот как будет выглядеть фрагмент параметров запроса, в котором собираются данные о количестве визитов (ym:s:visits) и просмотров страниц (ym:s:pageviews):

'metrics': 'ym:s:visits,ym:s:pageviews'

#### 2. Параметр dimensions

В параметре dimensions нужно перечислить список группировок (максимальное количество в одном запросе — 10).

Это необязательный параметр, и если не задать группировки, то в отчёте будет представлена общая статистика по сайту.

Если в запросе используется несколько группировок, то их, так же как и метрики, нужно перечислить в одну строку через запятую.

Группировки более разнообразны чем метрики в том плане, что обычно в запросах используется какой-то более-менее фиксированный набор метрик, а вот список группировок часто кардинально отличается в зависимости от выполняемой задачи.

Примеры группировок, необходимых для построения отчёта по поисковым системам с детализацией до запросов и посадочных страниц:

    ym:s:date — дата;
    ym:s:<attribution>SearchEngine> — поисковая система;
    ym:s:<attribution>SearchPhrase> — поисковый запрос;
    ym:s:startURL — посадочная страница

Если мы хотим сгруппировать в отчёте данные по дате и по посадочной странице, в параметрах необходимо оформить соответствующую часть таким образом:

'dimensions': 'ym:s:date,ym:s:startURL'

Мы видим, что в некоторых группировках содержится параметр <attribution>. Это означает, что при использовании таких группировок нужно задать атрибуцию трафика — информацию о том, как сессии пользователей будут привязываться к источнику трафика.

Например, если мы поставили атрибуцию по последнему источнику, то к конкретному источнику в отчёте будут отнесены сессии посетителей, у которых этот источник был последним.

Существует несколько типов атрибуции трафика:

* first — атрибуция по первому источнику;
* last — атрибуция по последнему источнику;
* prev — атрибуция по предыдущему источнику;
* lastSign — атрибуция по последнему значимому источнику.

Атрибуцию можно задать двумя способами:

1. Подставить вместо <attribution> один из типов (first, last или lastSign). Например:

'dimensions': 'ym:s:lastSignSearchEngine'

2. Оставить <attribution> в названии типа группировки и добавить в запрос параметр attribution. Например:

'dimensions': 'ym:s:<attribution>SearchEngine'
'attribution': 'lastSign'

#### 3. Параметр date1

Этот параметр содержит дату начала отчётного периода в формате YYYY-MM-DD (например, 2015-08-31). Кроме абсолютных значений в формате YYYY-MM-DD, можно задавать относительные значения:

    today — сегодняшняя дата;
    yesterday — вчерашняя дата;
    ndaysAgo — n дней назад от сегодняшней даты, где вместо n надо указать количество дней, например, 30daysAgo.
    
Если не задать начальную дату, то применится значение по умолчанию — 6daysAgo (6 дней назад от текущей даты).

В нашем тестовом запросе поставим в качестве даты начала отчёта 1 июня 2019 года:

'date1': '2019-06-01'

#### 4. Параметр date2

Дата окончания отчётного периода в формате YYYY-MM-DD (формат такой же, как и у параметра date1).

Важно! Значение параметра date2 не может быть меньше date1. 

Кстати, параметр date2 может относиться даже к будущему времени, запрос всё равно будет работать.

#### 5. Параметр limit

Параметр limit — это ограничение на количество возвращаемых результатов. По умолчанию он равен 100, а максимальное значение, которое можно использовать в запросе, — 100000. 

Значения limit=100000 хватает для большинства запросов, но надо понимать, что иногда отчёт может содержать большее количество строк и, чтобы получить все данные такого отчёта, может понадобиться параметр offset.

#### 6. Параметр offset
Параметр offset определяет номер первой строки выборки данных. Значение по умолчанию — 1, что означает, что отчёт будет генерироваться с первой строки, если параметр не задан.

Этот параметр полезен, когда полный набор данных в запрошенном отчёте содержит более 10000 строк.

Предположим, отчёт по интересующим нас параметрам содержит 25000 строк. Чтобы загрузить все данные этого отчёта, нам необходимо будет сделать несколько запросов к API Яндекс.Метрики. Первым запросом с offset=1 и  limit=10000 мы выгрузим первые 10000 строк. Установив offset=10001 мы выгрузим данные, начиная с 10001 до 20000 строки. В третьем запросе, установив offset=20001, мы загрузим оставшиеся данные. 

#### 7. Параметр ids
Параметр ids содержит перечень идентификаторов счётчиков. Это поле поддерживает более одного значения, что означает, что мы можем одним запросом выгружать данные с нескольких счётчиков.

Важно! Если мы включаем в отчёт данные более одного счётчика, то они будут объединены в общий отчёт, в котором не будет пометок о том, какие данные относятся к какому сайту.

Чтобы всё-таки получить отчёт по каждому сайту отдельно, можно передать в отчёт какой-то параметр, который будет точно идентифицировать каждый из сайтов, например добавить группировку Домен страницы входа (ym:s:startURLDomain).

### Работа со структурой отчёта

Yам нужно будет извлечь из каждого элемента списка metrika_data['data'] значение группировки (['dimensions'][0]['name']) и значения обеих метрик, после чего создать новый словарь, в котором ключами будут названия метрик и группировок, а значениями — извлечённые из старого отчёта данные.

```python
# Создаём пустой список, в который будем добавлять словари в новом формате
result = []
# Начинаем перебор элементов старого отчёта (только раздел с данными)
for data_item in metrika_data['data']:
    # Создаём словарь для хранения текущих данных в новом формате 
    new_dict = {}  
    # Обращаемся к разделу отчёта, содержащего метаданные
    # Перебираем названия группировок, которые использовались в отчёте
    for i,dimension in enumerate(data_item['dimensions']):
        # Создаём в новом словаре ключ для каждой группировки
        # Устанавливаем значение для каждого ключа
        new_dict[metrika_data['query']['dimensions'][i]] = dimension['name']
    # Те же действия выполняем для метрик
    for i,metric in enumerate(data_item['metrics']):
        new_dict[metrika_data['query']['metrics'][i]] = metric
    # Добавляем созданный словарь в итоговый список result
    result.append(new_dict)
result
```
